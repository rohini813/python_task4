```python
# Mobile Recharge Validation System
```

```python
def recharge_validation():

  valid_plan=[199, 299, 399, 599]

  while True:
    amount=int(input("Enter the amount: "))

    if amount< 50:
       print("Recharge amount must be at least 50rs.")
    elif amount not in valid_plan:
       print("Invalid plan. vaild plan are:",valid_plan)
    else:
      print("Recharge Successful")
      break

recharge_validation()
```

```python
#Inventory Reorder Alert System
```

```python
inventory={"Rice":20, "Oil":25, "Sugar":12, "Soap":30}

def check_inventory(inventory):
  for item, quantity in inventory.items():
    if quantity < 15:
      print(item,"-> Reorder Alert")
    else:
      print(item,"-> Stock OK")

check_inventory(inventory)
```

```python
#Student Result Processing System
```

```python
def result_processing(marks):
  total=0

  for mark in marks:
    total+=mark

  average=total/len(marks)

  print("Average Marks: " ,average)

  if average >= 50:
    return "Pass"
  else:
    return "Fail"

student_marks=[60, 55, 40, 70]
print(result_processing(student_marks))
```

```python
#Cab Fare Estimator with Retry Option
```

```python
def calculate_fare(distance, peak):
  base_fare=50
  fare = base_fare + (distance*12)

  if peak == "Yes":
    fare=fare*1.25

  return fare


while True :
  km =float(input("Enter distance in km: "))
  peak_hour = input("Is it peak hour? (Yes/No): ").lower()

  total_fear=calculate_fare(km, peak_hour)
  print("Total fare: ",total_fear)

  retry=input("Do you want to retry? (Yes/No): ").lower()
  if retry != "yes":
    break

```

```python
#Employee Attendance Eligibility Checker
```

```python
def attendance_check(attendance_list):
  total_days=len(attendance_list)
  present_days=0

  for attendance in attendance_list:
    if attendance == "P":
      present_days+=1

  percentage = (present_days/total_days)*100

  print("Attendance Percentage",percentage)

  if percentage >= 75:
    return "Eligible for Promotion"
  else:
    return "Not Eligible"

attendance=["P", "P", "P", "P", "P", "A", "P"]
print(attendance_check(attendance))
```

```python
#Password Strength Checker
```

```python
def password_checker(password):
    special_char="@#$"
    has_digit=False
    has_special=False

    if len(password) < 8:
      return "Weak Password X (Minimum 8 characters required)"

    for char in password:
      if char.isdigit():
         has_digit=True
      if char in special_char:
         has_special=True


    if has_digit and has_special:
     return "Strong Password"

    else:
      return "Weak Password X (Must contain digit & special character)"

user_password=input("Enter the password: ")
print(password_checker(user_password))
```
